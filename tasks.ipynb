```python
from bs4 import *
import requests, re

def animal_species():
	url = 'https://inlnk.ru/jElywR'
	base_url = 'https://ru.wikipedia.org/'

	letters = {}

	while True:
		soup = BeautifulSoup(requests.get(url).text, 'html.parser')
		for cat in soup.find_all(class_='mw-category mw-category-columns'):
			for subcat in cat.find_all(class_='mw-category-group'):
				if re.search(r'[А-Я]', subcat.find('h3').text):
					key = subcat.find('h3').text
					if key in letters:
						letters[key] += len(subcat.find_all('li'))
					else:
						letters[key] = 0
						letters[key] += len(subcat.find_all('li'))
					url = base_url + soup.find_all('a', string='Следующая страница', limit=1)[-1]['href']
				else:
					return letters

for letter, amount in animal_species().items():
    print(letter, ':', amount)
```
